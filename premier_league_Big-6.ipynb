```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Load the data from csv file
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
# Finding the club's overall
Mun = fifa.loc[fifa['Club'] == 'Manchester United']['Overall']
Che = fifa.loc[fifa['Club'] == 'Chelsea']['Overall']
Liv = fifa.loc[fifa['Club'] == 'Liverpool']['Overall']
Mci = fifa.loc[fifa['Club'] == 'Manchester City']['Overall']
Ars = fifa.loc[fifa['Club'] == 'Arsenal']['Overall']
Tot = fifa.loc[fifa['Club'] == 'Tottenham Hotspur']['Overall']

# Overall's of first ten Man U players
Mun[:10]
```

```python
plt.figure(figsize=(10,13))

# inputs grouped as list
clubs = [Mun,Che,Liv,Mci,Ars,Tot]
# labelling the inputs
labels = ['Manchester United','Chelsea', 'Liverpool', 'Manchester City', 'Arsenal', 'Tottenham']
# Title of the chart
plt.title('Premier League Big Six',fontdict={'fontsize':20, 'fontweight':'bold'})
# Labelling of the Y-Axis
plt.ylabel('FIFA 2018 Overall Rating',fontdict={'fontsize':10, 'fontweight':'bold'})

# Plotting the box and whiskers chart
boxes = plt.boxplot(clubs, labels=labels, medianprops={'linewidth':2,'color':'r'}, patch_artist=True)

for box in boxes['boxes']:
    box.set(color='#4286f4',facecolor = '#e0e0e0',hatch='/', linewidth=2)

plt.show()
```
